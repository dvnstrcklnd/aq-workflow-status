```python
session = get_session("production")
```

```python
columns = [
    'Name',
    "Description",
    'Created',
    'Project',
    'Manufacturer', 
    'Oligo Library ID', 
    'min length (nt) (array)', 
    'max length (nt) (array)', 
    'variants (array)', 
    'sublibrary name (array)', 
    'forward priming site (array)', 
    'reverse priming site (array)', 
    'inner forward primer (array)', 
    'inner reverse primer (array)', 
    'sublibrary forward primer (array)', 
    'sublibrary reverse primer (array)'
]
```

```python
alt_columns = [
    "Timestamp",
    "Email Address",
    "Your Name",
    "Library Name",
    "Description",
    "How is (or was) the library provided to us?",
    "Approximately when did (or will) we receive the library?",
    "Oligo Manufacturer",
    "Oligo Library ID",
    "Min Length (nt)",
    "Max Length (nt)",
    "Variants",
    "Forward Priming Site",
    "Reverse Priming Site",
    "Plasmid",
    "Cloning site",
    "Background Strain",
    "Selectable markers",
    "Induction",
    "Project"
]
```

```python
oligo_pool_ot = session.SampleType.find_by_name("Oligo Pool")
oligo_pool_samples = oligo_pool_ot.samples
```

```python
fts = oligo_pool_ot.field_types
ftnames = [ft.name for ft in fts]

data = []
for s in oligo_pool_samples:
    d = {}
    d['Name'] = s.name
    d['Created'] = s.created_at
    d["Description"] = s.description
    d['Project'] = s.project
    
    for n in ftnames:
        d[n] = s.properties.get(n)
        if hasattr(d[n], 'name'):
            d[n] = d[n].id
        
    data.append(d)
```

```python
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
year = 2019
month = 8

if month < 12:
    next_month = month + 1
    next_year = year
else:
    next_month = 1
    next_year = year + 1
    
pacific = timezone('US/Pacific')
# central = timezone('US/Central')

month_start = pacific.localize(datetime(year, month, 1))
month_end = pacific.localize(datetime(next_year, next_month, 1))
```

```python
ft_names = [
    "Manufacturer",
    "Oligo Library ID",
    "min length (nt) (array)",
    "max length (nt) (array)",
    "variants (array)",
    "sublibrary name (array)",
    "forward priming site (array)",
    "reverse priming site (array)"
]   
alt_data = []

for s in oligo_pool_samples:
    properties = [
        s.properties.get("sublibrary name (array)"),
        s.properties.get("min length (nt) (array)"),
        s.properties.get("max length (nt) (array)"),
        s.properties.get("variants (array)"),
        s.properties.get("forward priming site (array)"),
        s.properties.get("reverse priming site (array)")
    ]
    
    if isinstance(properties[0], list):
        properties = list(map(list, zip(*properties)))
    else:
        properties = [properties]
    
    print("{} {}".format(s.id, properties))
    
    continue
        
    for i, p in enumerate(properties):
        ad = {}
        ad["Library Name"] = p[0]
        ad["Timestamp"] = s.created_at
        
        dna_library = session.Sample.find_by_name(ad["Library Name"])
        if dna_library:
            ad["Description"] = dna_library.description
        else:
            ad["Description"] = s.description
            
        ad['Project'] = s.project
        ad["Oligo Manufacturer"] = s.properties.get("Manufacturer")
        ad["Oligo Library ID"] = s.properties.get("Oligo Library ID")
        
        ad["Min Length (nt)"] = p[1]
        ad["Max Length (nt)"] = p[2]
        ad["Variants"] = p[3]
        ad["Forward Priming Site"] = p[4]
        ad["Reverse Priming Site"] = p[5]

        alt_data.append(ad)
        
alt_df = pd.DataFrame(data=alt_data, columns=alt_columns)
alt_df
```

```python
alt_df.to_csv("oligo_pools.csv", index=False)
```
